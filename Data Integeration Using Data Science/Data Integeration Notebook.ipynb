```python
'''
PANDAS LIRARY IS USED TO MANIPULATE THE DATA 
'''

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
'''
READING TSV FILE 
'''
fitness_pal_data = pd.read_csv('/data/workspace_files/data.tsv', delimiter='\t')
```

```python
'''
WE HAVE TO SEPERATE FOOD IDS AND WRITE THEM IN A SEPERATE ROW THATS WHY I 
WRITE THIS FUNCTION EXPLODE TO WRITE 1,2,3,4 FOOD IDS INTO SEPERATE ROW
'''

fitness_pal_data['food_ids'] = fitness_pal_data['food_ids'].str.split(',')
fitness_pal_data = fitness_pal_data.explode('food_ids')

# Reset the index
fitness_pal_data = fitness_pal_data.reset_index(drop=True)
```

```python

'''
NOW WE CAN SEEN THAT FOOD IDS ARE SAPERATED 
'''
fitness_pal_data.head(3)
```

```python


'''
I WILL COMBINE FITNESS PAL ITEM WITH MEAL_ID BY USING INNER JOIN 
'''
fitness_pal_item = pd.read_csv('/data/workspace_files/items.tsv', delimiter='\t')
```

```python
'''
SHOWING FITNESS PAL ITEM CONTAIN ID AND ITEMS
'''

fitness_pal_item.head(2)
```

```python
'''
I APPLIED INNER JOIN TO COMBINE FITNESS PAL DATA AND FITNESS PAL ITEM 
BASED ON ID FITNESS PAL DATA HAVE MEAL ID AND FITNESS PAL ITEM HAVE ID
ON THE BASIS OF THAT I APPLIED JOIN TO COMBINE THE DATASET 
'''

fitness_combined_data = pd.merge(fitness_pal_data, fitness_pal_item, left_on='meal_id', right_on='id', how='inner')
```

```python
'''
SHOWING FITNESS COMBINED DATA 
'''
fitness_combined_data.head(2)
```

```python
'''
READING USDA DATA 
'''
usda = pd.read_csv('/data/workspace_files/USDA.csv')
```

```python
'''
SHOWING USDA DATA
'''
usda.head(2)
```

```python
'''
CONVERT BOTHE COLUMNS INTO SAME INT 64 FORMAT SO WE CAN APPLY MERGE COLUMN
'''

usda['ID'] = usda['ID'].astype('int64')
fitness_combined_data['food_ids'] = fitness_combined_data['food_ids'].astype('int64')
```

```python
'''
I APPLIED INNER JOIN TO COMBINE USDA AND FITNESS COMBINED DATA
BASED ON ID USDA  HAVE  ID AND FITNESS COMBINED DATA HAVE FOOD IDS
ON THE BASIS OF THAT I APPLIED JOIN TO COMBINE THE DATASET

'''
Merge_data = pd.merge(usda, fitness_combined_data, left_on='ID', right_on='food_ids', how='inner')
```

```python
'''
SHOWING MERGE DATA
'''
Merge_data.head(2000)
```

```python
'''
WE WILL GET THE USEFUL COLUMNS TO PERFORM OUR ANALYSIS 
'''

Merge_data=Merge_data[['date','Description','item','Calories','Protein','TotalFat','Carbohydrate','Sodium','SaturatedFat','Cholesterol','Sugar','Calcium','Iron','Potassium','VitaminC','VitaminE','VitaminD']]
```

```python
'''
SHOWING RE ARRAGING MERGE DATA 
'''
Merge_data.head(2)
```

```python
'''
CONVERT DATE INTO STANDARD FORMAT WHICH IS DATETIME FORMAT
'''
Merge_data['date'] = pd.to_datetime(Merge_data['date'])
```

```python
'''
GET YEAR , MONTH , DAY FROM DATE TIME 
'''
# Extract year, month, and day into separate columns
Merge_data['year'] = Merge_data['date'].dt.year
Merge_data['month'] = Merge_data['date'].dt.month
Merge_data['day'] = Merge_data['date'].dt.day
```

```python
'''
RE ARRANGING COLUMNS 
'''
Merge_data=Merge_data[['date','year','month','day','Description','item','Calories','Protein','TotalFat','Carbohydrate','Sodium','SaturatedFat','Cholesterol','Sugar','Calcium','Iron','Potassium','VitaminC','VitaminE','VitaminD']]
```

```python
'''
REMOVE DUPLICATES ROWS IN THE DATASET
'''
Merge_data.drop_duplicates(inplace=True)
```

```python
null_percentage = (Merge_data.isnull().mean() * 100).round(2)
null_percentage_df = pd.DataFrame({'Column': null_percentage.index, 'NullPercentage': null_percentage.values})
null_percentage_df.head(20)
```

```python
'''
CHECKING THE PERCENTAGE OF OUTLIERS 
'''

# Set the threshold for identifying outliers (e.g., 1.5 times the IQR)
outlier_threshold = 1.5

# Calculate the percentage of outliers in each column
outlier_percentages = []
for column in Merge_data.columns:
    if np.issubdtype(Merge_data[column].dtype, np.number):  # Check if column is numerical
        # Calculate the quartiles and IQR
        q1 = Merge_data[column].quantile(0.25)
        q3 = Merge_data[column].quantile(0.75)
        iqr = q3 - q1

        # Determine the outlier range
        lower_range = q1 - outlier_threshold * iqr
        upper_range = q3 + outlier_threshold * iqr

        # Count the number of outliers
        outliers = Merge_data[(Merge_data[column] < lower_range) | (Merge_data[column] > upper_range)]
        outlier_percentage = (len(outliers) / len(Merge_data)) * 100
        outlier_percentages.append(outlier_percentage)
    else:
        outlier_percentages.append(0)

# Create a DataFrame with column names and outlier percentages
result_df = pd.DataFrame({'Column': Merge_data.columns, 'OutlierPercentage': outlier_percentages})
```

```python
'''
SHOWING PERCENTAGE OF OUTLIERS
'''
result_df.head(25)
```

```python
'''
GETTING LIST FROM COLUMN NAMES TO CHECK OULIERS
'''

columns_to_check=list(Merge_data.columns)
```

```python
'''
THIS CODE BELOW CHECK THE NUMBER OF OULIERS AND NULL VALUES PRESENT IN SPECIFIC COLUMN OR NOT
'''
# Set the threshold for identifying outliers (e.g., 1.5 times the IQR)
outlier_threshold = 1.5


# Check for outliers and null values in the specified columns
for column in columns_to_check:
    if np.issubdtype(Merge_data[column].dtype, np.number):  # Check if column is numerical
        # Calculate the quartiles and IQR
        q1 = Merge_data[column].quantile(0.25)
        q3 = Merge_data[column].quantile(0.75)
        iqr = q3 - q1
        
        # Determine the outlier range
        lower_range = q1 - outlier_threshold * iqr
        upper_range = q3 + outlier_threshold * iqr
        
        # Check for outliers
        outliers = Merge_data[(Merge_data[column] < lower_range) | (Merge_data[column] > upper_range)]
        
        # Check for null values
        null_values = Merge_data[Merge_data[column].isnull()]
        
        if len(outliers) > 0:
            print(f"Outliers found in {column}:")
        else:
            print(f"No outliers found in {column}")
        
        if len(null_values) > 0:
            print(f"Null values found in {column}:")
        else:
            print(f"No null values found in {column}")
    else:
        print(f"{column} is not a numerical column")
```

```python
'''
FILLING NULL VALUES IF THE NUMERICAL COLUMN VALUE CONTAIN OUTLIER WE FILL IT 
WITH MEDIAN VALUE IF IT DONOT CONTAIN OUTLIER WE FILL IT WITH MEAN VALUE
'''
# Set the threshold for identifying outliers (e.g., 1.5 times the IQR)
outlier_threshold = 1.5

# Iterate over each column in the DataFrame
for column in Merge_data.columns:
    if np.issubdtype(Merge_data[column].dtype, np.number):  # Check if column is numerical
        # Calculate the quartiles and IQR
        q1 = Merge_data[column].quantile(0.25)
        q3 = Merge_data[column].quantile(0.75)
        iqr = q3 - q1

        # Determine the outlier range
        lower_range = q1 - outlier_threshold * iqr
        upper_range = q3 + outlier_threshold * iqr

        # Check if outliers and null values are present
        has_outliers = any((Merge_data[column] < lower_range) | (Merge_data[column] > upper_range))
        has_nulls = Merge_data[column].isnull().any()

        if has_outliers and has_nulls:
            # Fill null values with median
            Merge_data[column].fillna(Merge_data[column].median(), inplace=True)
        elif not has_outliers and has_nulls:
            # Fill null values with mean
            Merge_data[column].fillna(Merge_data[column].mean(), inplace=True)
    else:
        continue
```

```python
'''
MIN MAX THE DATA BETWEEN 0 AND 1 THE PURPOSE OF THIS IS TO
STANDARDIZE THE DATA
'''


# Select numerical columns
numerical_columns = Merge_data.select_dtypes(include='number').columns

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the numerical data using MinMaxScaler
Merge_data[numerical_columns] = scaler.fit_transform(Merge_data[numerical_columns])
```
